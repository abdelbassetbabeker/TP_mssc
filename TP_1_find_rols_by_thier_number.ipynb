```python
import functools



rolenumber =int(input("Please Type Role Number : "));
co = 0

GenerationsLimit = 50

size = 50

FirstGeneration = [0]*size
FirstGeneration[24]=1

TempGeneration = [0]*size

GenerationEvalution=[]
GenerationEvalution.append(FirstGeneration)


def decimalToBinary(n):
    return bin(n).replace("0b", "")

onezesBin=[]
onezesDec=[]

for i in range(7 , 0 , -1):
    if 2**i <= rolenumber:
        if (2**i + co) <= rolenumber:
            co = co + 2**i
            onezesBin.append(decimalToBinary(i))
            onezesDec.append(i)


def testRole(x,y,z):
    tupleTest = x,y,z
    n1 = functools.reduce(lambda sub, ele: sub * 10 + ele, tupleTest)
    n =str(n1)
    if(n in onezesBin):
        return 1
    else :
        return 0




for i in range(GenerationsLimit):
    for j in range(size):
        TempGeneration[(j) % size] = testRole(FirstGeneration[(j-1) % size] , FirstGeneration[(j) % size] , FirstGeneration[ (j+1) % size])
    FirstGeneration = TempGeneration.copy()
    GenerationEvalution.append(FirstGeneration)



print('Powers : ',onezesDec)
print('where ones :' , onezesBin)












## displaying the result in ghraphical way

import tkinter as tk


class Level:
    def __init__(self, name, height, width):
        self.name = name
        self.height = height
        self.width = width
        self.tiles = makeTiles(height, width)

    class Tile:
        def __init__(self, x, y, type, color, isOccupied, enemy):
            self.x = x
            self.y = y
            self.type = type
            self.color = color
            self.isOccupied = isOccupied
            self.enemy = enemy

def makeLevel(name, height, width):
    level = Level(name, height, width)
    return level

def makeTiles(height, width):
    tiles = [[Level.Tile(x, y, "Surface", "red", False, "None") for x in range(width)] for y in range(height)]
    return tiles

test1 = makeLevel("test", 50, 50)
for x in range(GenerationsLimit):
    for y in range(size):
        if GenerationEvalution[x][y] == 0:
            test1.tiles[x][y].color = "white"
        elif GenerationEvalution[x][y] == 1:
            test1.tiles[x][y].color = "black"

colorMatrix = [[0 for x in range(50)] for y in range(50)]
for x in range(50):
    for y in range(50):
        colorMatrix[x][y] = test1.tiles[x][y].color

#print(colorMatrix)

width, height = 600, 600

root = tk.Tk()
root.title("colorMatrix")

frame = tk.Frame()
frame.pack()

canvas = tk.Canvas(frame, width=width, height=height)
rows, cols = len(colorMatrix), len(colorMatrix[0])

rect_width, rect_height = width // rows, height // cols
for y, row in enumerate(colorMatrix):
    for x, color in enumerate(row):
        x0, y0 = x * rect_width, y * rect_height
        x1, y1 = x0 + rect_width-1, y0 + rect_height-1
        canvas.create_rectangle(x0, y0, x1, y1, fill=color, width=0)
canvas.pack()

root.mainloop()
```
